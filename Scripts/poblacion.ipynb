```python
import pandas as pd
import os
```

```python
ruta_a = r"C:\Users\Hp\Downloads\poblacion_municipal_2005_2017.csv"
ruta_b = r"C:\Users\Hp\Downloads\poblacion_municipal_2018_2042.csv"

df_a = pd.read_csv(ruta_a, encoding='utf-8', low_memory=False)
df_b = pd.read_csv(ruta_b, encoding='utf-8', low_memory=False)


df = pd.concat([df_a, df_b], ignore_index=True)
```

```python
df_filtrado = df[(df["AÑO"] >= 2017) & (df["AÑO"] <= 2023)]
df_filtrado = df_filtrado[df_filtrado["ÁREA GEOGRÁFICA"] == "Total"]
df_filtrado = df_filtrado[df_filtrado["DPNOM"] == 'Atlántico']

print(df_filtrado["AÑO"].unique())
```

```python
df_filtrado.info()
```

```python
df_filtrado['MUNICIPIO'] = df_filtrado['MUNICIPIO'].str.upper()

df_filtrado['MUNICIPIO'].unique()
```

```python
df_filtrado.to_csv("Poblacion_atl.csv", encoding='latin_1')
```

```python
df_pob = pd.read_csv(r"C:\Users\Hp\DENGUE\scripts\Poblacion_atl.csv", encoding='latin_1', low_memory=False)
df_pob.info()
```

```python
tabla_wide = (
    df_pob.pivot_table(
        index="MUNICIPIO",       # lo que queda como fila
        columns="AÑO",           # cada año será una columna
        values="Población",      # lo que se distribuye
        aggfunc="sum",           # suma (por si hay más de un registro)
        fill_value=0             # si falta valor, poner 0
    )
    .reset_index()
)

# Agregar columna con total por municipio
tabla_wide["TOTAL"] = tabla_wide.drop(columns="MUNICIPIO").sum(axis=1)

tabla_wide
```

```python
tabla_wide.to_csv("tabla_municipios.csv", index=False, encoding="utf-8-sig")
```
